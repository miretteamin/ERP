```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

As 0.5 second = 50 samples and each trial should be 2 seconds = 50*4 = 200 samples

```python
# Class One
# signal1 = 2 * np.sin(2 * np.pi * 4 * time)
signal1 = np.random.uniform(0.0, 2.0, 250*50)

# Split to trials
trials1 = np.array(np.array_split(signal1, 50)) # (50, 40/20
```

```python
# Class Two
signal2 = np.random.uniform(0.0, 2.0, 250*50)

# Split to trials
trials2 = np.array(np.array_split(signal2, 50)) # (50, 40/20
```

```python
plt.plot(trials1[0])
plt.plot(trials2[0])
```

```python
def ERP(trials):
    # Stimulus Locked 

    # # Baseline Correction
    standardizedTrials = []
    for trial in trials:
        prestimulus = trial[0:50]
        poststimulus = trial[50:]

        prestimulusMean = np.mean(prestimulus)
        res = poststimulus - prestimulusMean
        standardizedTrials.append(res)
        
    standardizedTrials = np.array(standardizedTrials)
    # Averaging Trials
    avgTrials = np.mean(standardizedTrials, axis=0)

    return avgTrials
```

```python
res1 = ERP(trials1)
res2 = ERP(trials2)
```

```python
plt.plot(res1)
plt.plot(res2)
```
